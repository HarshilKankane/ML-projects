```python
data_new=data.drop(['fixed acidity','residual sugar','chlorides','free sulfur dioxide','density','total sulfur dioxide','citric acid','pH'],axis=1)
data_new
```

```python
x1=data_new['volatile acidity'].values
x2=data_new['sulphates']
x3=data_new['alcohol']
x1=x1.reshape(1599,1)
x2=x2.values.reshape(1599,1)
x3=x3.values.reshape(1599,1)
```

```python
from sklearn.model_selection import train_test_split
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y,test_size=0.2,random_state=0)
x2_train,x2_test,y2_train,y2_test = train_test_split(x2,y,test_size=0.2,random_state=0)
x3_train,x3_test,y3_train,y3_test = train_test_split(x3,y,test_size=0.2,random_state=0)
```

```python
Lin1.fit(x1_train,y1_train)
Lin2.fit(x2_train,y2_train)
Lin3.fit(x3_train,y3_train)
```

```python
y1_pred=Lin1.predict(x1_test)
y2_pred=Lin2.predict(x2_test)
y3_pred=Lin3.predict(x3_test)
```

```python
import matplotlib.pyplot as plt
plt.style.use('default')
plt.style.use('ggplot')
plt.plot(x1_test,y1_pred)
plt.plot(x2_test,y2_pred)
plt.plot(x3_test,y3_pred)
plt.scatter(x1_test,y1_pred)
plt.scatter(x2_test,y2_pred)
plt.scatter(x3_test,y3_pred)
plt.xlabel('Wine features')
plt.ylabel('Quality')

plt.show
```
